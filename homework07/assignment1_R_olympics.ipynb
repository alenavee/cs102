__1. How old were the youngest male and female participants of the 1996 Olympics?__

- 16 and 15
- **14 and 12 **
- 16 and 12
- 13 and 11

```python
min(df[df$Year == 1996 & df$Sex == 'M',]['Age'], na.rm = T)
min(df[df$Year == 1996 & df$Sex == 'F',]['Age'], na.rm = T)
```

__2. What was the percentage of male gymnasts among all the male participants of the 2000 Olympics? Round the answer to the first decimal.__

*Hint:* here and further if needed drop duplicated sportsmen to count only unique ones. 

- 0.2
- **1.5 **
- 2.5
- 7.7

```python
df_n <- df[!duplicated(df$Name),]
gymnasts <- df_n[df_n['Sport'] == 'Gymnastics' & df_n['Games'] == '2000 Summer' & df_n['Sex'] == 'M', ]
male_participants <- df_n[df_n['Games'] == '2000 Summer' & df_n['Sex'] == 'M',]
cat(nrow(gymnasts) / nrow(male_participants) * 100)
```

__3. What are the mean and standard deviation of height for female basketball players participated in the 2000 Olympics? Round the answer to the first decimal.__

- 178.5 and 7.2
- 179.4 and 10
- 180.7 and 6.7
- **182.4 and 9.1** 

```python
f_bb <- df[df$Year == 2000 & df$Sex == 'F' & df$Sport == 'Basketball',]
f_bb <- f_bb[!duplicated(f_bb$ID),]
round(mean(f_bb$Height), digits=1) 
round(sd(f_bb$Height), digits=1)
```

__7. Is it true that Switzerland won fewer medals than Serbia at the 2016 Olympics? Do not consider NaN values in _Medal_ column.__ 

- **Yes** 
- No

```python
switz <- nrow(df[df['Team'] == 'Switzerland' & !is.na(df['Medal']) & df['Games'] == '2016 Summer', ])
serbia <- nrow(df[df['Team'] == 'Serbia' & !is.na(df['Medal']) & df['Games'] == '2016 Summer', ])
switz < serbia
```

__8. What age category did the fewest and the most participants of the 2014 Olympics belong to?__
- [45-55] and [25-35) correspondingly
- **[45-55] and [15-25) correspondingly**
- [35-45] and [25-35) correspondingly
- [45-55] and [35-45) correspondingly

```python
oldest <- df[(55 >= df['Age']) & (df['Age'] >= 45), ]
adults <- df[(45 >= df['Age']) & (df['Age'] >= 35), ]
medium <- df[(35 > df['Age']) & (df['Age'] >= 25),]
youngest <- df[(25 > df['Age']) & (df['Age'] >= 15), ]
cat('oldest(45-55):', nrow(oldest),'\n')
cat('adults(35-45):', nrow(adults),'\n')
cat('medium(25-35):', nrow(medium),'\n')
cat('youngest(15-25):', nrow(youngest),'\n')
```

__9. Is it true that there were Summer Olympics held in Lake Placid? Is it true that there were Winter Olympics held in Sankt Moritz?__

- Yes, Yes
- Yes, No
- **No, Yes** 
- No, No 

```python
lake_placid <- nrow(df[(df['City'] == 'Lake Placid') & (df$Season == 'Summer'),]) > 0
sankt_moritz <- nrow(df[(df['City'] == 'Sankt Moritz') & (df$Season == 'Winter'),]) > 0
cat(lake_placid, '\n')
cat(sankt_moritz)
```
